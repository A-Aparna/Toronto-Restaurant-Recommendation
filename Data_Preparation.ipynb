```python
#Introducing a feature named cuisine
df['cuisine']='NA'
df.head()
```

```python
#alloting cuisines to the restaurants based on the details in Categories column
def cuisine_type(row):
    words=np.array(['Mexican','Indian','Japan','Italian','Middle','Thai','Chinese','Mediterranean','Canadian',\
                    'Greek','Peruvian','Breakfast & Brunch','Sushi','Baker','Juice','Barbeque','Korean','French',\
                   'Vietnamese'])
    for word in words:
        if (row.find(word) != -1) :
            return word
        
    return None
    
df['cuisine']=df['categories'].apply(cuisine_type)
```

```python
#Condesing the cuisines to 10 categories but also retaining the rows
def condense_cuisine(row):
    if (row=='Sushi' or row=='Korean'):
        return 'Japan'
    if (row=='Vietnamese' ):
        return 'Thai'
    if (row=='French' or row=='Italian'):
        return 'European'
    if (row=='Greek' or row=='Middle'):
        return 'Mediterranean'
    if (row=='Baker' or row=='Juice'):
        return 'Dessert'
    if (row=='Barbeque'):
        return 'Canadian'
    else:
        return row
    return None
df['cuisine']=df['cuisine'].apply(condense_cuisine)
```
